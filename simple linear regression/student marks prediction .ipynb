```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import numpy as np
```

```python
x=df[["study_hours"]]
y=df[["student_marks"]]
```

```python
plt.figure(figsize=(7,5))
plt.scatter(x=df.study_hours,y=df.student_marks,color="red",marker="*",s=100,alpha=0.5)
plt.xlabel("study_hours")
plt.ylabel("student_marks")
plt.title("study_hours Vs student_marks")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(df.corr(),annot=True,linewidths=3)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_text=train_test_split(x,y,test_size=0.2,random_state=51)
```

```python
x_test.shape 
```

```python
y_text.shape
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
y_pred
```

```python
pd.DataFrame(np.c_[x_test,y_text,y_pred],columns=["study_hours","student_marks_original Percentage","student_marks_predicted Percentage"])
```

```python
import joblib
joblib.dump(lr, "Student_mark_prediction_model.pkl")
```
